```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020 = pd.read_csv("Resources/athletic_sales_2020.csv")
sales_2021 = pd.read_csv("Resources/athletic_sales_2020.csv")
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
joined_athletic_sales = pd.concat([sales_2020, sales_2021], axis=0, join="inner").reset_index(drop=True) 
```

```python
# Check if any values are null.
joined_athletic_sales.isna().sum()
```

```python
# Check the data type of each column
joined_athletic_sales.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
joined_athletic_sales['invoice_date'] = pd.to_datetime(joined_athletic_sales['invoice_date'])
joined_athletic_sales.head()
```

```python
# Confirm that the "invoice_date" data type has been changed.
joined_athletic_sales.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
total_products_sold = joined_athletic_sales.groupby(['region', 'state', 'city'])\
        .agg(Total_Products_Sold=('units_sold', 'sum'))

# Show the top 5 results.
total_products_sold.sort_values(by=(['Total_Products_Sold']), ascending=False).head(5)                            
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
total_products_sold = pd.pivot_table(joined_athletic_sales,
                                     index=['region', 'state', 'city'],
                                     values='units_sold',
                                     aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"
total_products = total_products_sold.rename(columns={"units_sold": "Total_Products_Sold"})

# Show the top 5 results.
total_products.sort_values(by=(["Total_Products_Sold"]), ascending=False).head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
women_athletic_footwear = joined_athletic_sales[joined_athletic_sales["product"].str.contains("women's athletic footwear", case=False, regex=True)]
women_athletic_footwear.head()
```

```python
women_athletic_footwear.info()
```

```python
women_athletic_footwear.isna().sum()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
women_athletic_footwear_sold = women_athletic_footwear.groupby(['retailer', 'region', 'state', 'city'])\
    ['units_sold'].agg(women_athletic_footwear_sold=("sum"))
# Show the top 5 results.
women_athletic_footwear_sold.sort_values(by=(['women_athletic_footwear_sold']), ascending=False).head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
women_athletic_footwear_sold = pd.pivot_table(women_athletic_footwear,
                                              values='units_sold',
                                              index=['retailer', 'region', 'state', 'city'],
                                              aggfunc='sum')

women_athletic_footwear_sold
```

```python

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
women_athletic_footwear_sold = women_athletic_footwear_sold.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})

```

```python
# Show the top 5 results.
top_5_women_athletic_footwear_sold = women_athletic_footwear_sold.sort_values(by="Womens_Footwear_Units_Sold", ascending=False).head(5)

top_5_women_athletic_footwear_sold
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column as the index, and the "total_sales" as the values.
women_sales_footwear = pd.pivot_table(women_athletic_footwear, index=['invoice_date'],
                                      values='total_sales',
                                      aggfunc='sum')

women_sales_footwear
```

```python

# Optional: Rename the "total_sales" column to "Total Sales"
women_athletic_footwear_sold = women_athletic_footwear_sold.rename(columns={"total_sales": "Total Sales"})

# Show the table.
women_athletic_footwear_sold.head(10)
```

```python
# Assuming 'invoice_date' is in datetime format, if not, convert it
women_athletic_footwear['invoice_date'] = pd.to_datetime(women_athletic_footwear['invoice_date'])

# Create a pivot table with the 'invoice_date' column as the index, and the "total_sales" as the values.
women_sales_footwear = pd.pivot_table(women_athletic_footwear, index=['invoice_date'],
                                      values='total_sales',
                                      aggfunc='sum')

# Resample the pivot table into daily bins, and get the total sales for each day.
daily_womans_footwear_sales = women_sales_footwear.resample('D').sum()

# Sort the resampled pivot table in descending order on "total_sales".
top_10_daily_womans_footwear_sales = daily_womans_footwear_sales.sort_values(by="total_sales", ascending=False).head(10)

top_10_daily_womans_footwear_sales
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_womens_footwear_sales = women_sales_footwear.resample('W').sum()

# Sort the resampled pivot table in descending order on "total_sales".
top_10_weekly_womens_footwear_sales = weekly_womens_footwear_sales.sort_values(by="total_sales", ascending=False).head(10)

top_10_weekly_womens_footwear_sales
```
